```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd


df=pd.read_csv("C:/Users/User/Downloads/H05.csv")
df.head()
```

1.Histogram distribusi nilai matematika.

```python
df[["nilai_mat"]].plot(kind="hist", bins=[0, 20, 40, 60, 80, 100], rwidth=0.8,color='pink', title="Histogram distribusi nilai matematika")
# Menampilkan plot
plt.show()
```

Barplot diatas menunjukkan bahwa sebagian besar mahasiswa memiliki nilai matematika di rentang 60-80.

2.Diagram batang horizontal banyaknya mahasiswa masing-masing fakultas.

```python
plt.figure(figsize=(8, 5))
df['fakultas'].value_counts().plot(kind='barh', color="pink", edgecolor="black")
plt.title('Jumlah Mahasiswa per Fakultas')
plt.xlabel('Jumlah Mahasiswa')
plt.ylabel('Fakultas')
plt.show()
```

3.Diagram pie banyaknya mahasiswa masing-masing fakultas

```python
plt.figure(figsize=(8, 8))
df['fakultas'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Presentase Mahasiswa per Fakultas')
plt.show()
```

4. Berdasarkan diagram pie dan batang horizontal manakah fakultas dengan mahasiswa terbanyak? manakan diagram yang lebih baik dalam menampilkan fakultas dengan mahasiswa terbanyak dan mengapa?

Diagram batang adalah pilihan yang lebih efektif karena kemampuannya dalam membandingkan nilai antar kategori secara visual. Meskipun diagram lingkaran dapat berguna untuk melihat proporsi, diagram batang lebih disarankan ketika fokus utama adalah pada pembacaan jumlah absolut.

5. Scatter plot dengan nilai kimia sebagai x dan fisika sebagai y.

```python
plt.figure(figsize=(8, 6))
plt.scatter(df['nilai_kim'], df['nilai_fis'], alpha=0.7, c='red', edgecolor='black')
plt.title('Scatter Plot Nilai Kimia vs. Nilai Fisika', fontsize=12)
plt.xlabel('Nilai Kimia', fontsize=10)
plt.ylabel('Nilai Fisika', fontsize=10)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

keterangan:Menunjukkan hubungan positif antara nilai kimia (sumbu x) dan fisika (sumbu y).
